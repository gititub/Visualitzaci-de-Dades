

L'exposició a PM2,5 es refereix a la inhalació de partícules fines que tenen un diàmetre de 2,5 micròmetres o menys. Aquestes partícules són prou petites per penetrar profundament en el sistema respiratori, arribar als pulmons i potencialment entrar al torrent sanguini. PM2.5 és una barreja de partícules diminutes i gotes líquides que poden incloure una varietat de components com ara àcids (per exemple, nitrats i sulfats), productes químics orgànics, metalls i partícules de sòl o pols.


Les partícules fines (PM2,5) són el contaminant atmosfèric que suposa un major risc per a la salut a nivell mundial, afectant més persones que qualsevol altre contaminant. L'exposició crònica augmenta considerablement el risc de patir malalties respiratòries i cardiovasculars (OMS, 2018). Per aquests motius, s'ha seleccionat l'exposició de la població a PM2,5 (exterior o ambiental) com a **indicador principal de creixement verd de l'OCDE.**

## Dades Actuals - 2020

```python
OECD = pd.read_csv('original_data/OECD.PM25.2020.csv')

pm25oecd = OECD[['Country', 'COUNTRY', 'TIME_PERIOD', 'OBS_VALUE']].rename(
    columns={'COUNTRY': 'Code', 'Country': 'Entity', 'TIME_PERIOD': 'Year', 'OBS_VALUE': 'PM2.5'}
)
pm25oecd.describe() #fins 2020
```

```python
pm25oecd.dtypes
```

```python
death_rate = pd.read_csv('original_data/OECD.MORT_CRUDE_RATIO.2021.csv')

death_rate = death_rate[['Country', 'COUNTRY', 'TIME_PERIOD', 'OBS_VALUE']].rename(
    columns={'COUNTRY': 'Code', 'Country': 'Entity', 'TIME_PERIOD': 'Year', 'OBS_VALUE': 'Death.Rate'}
)
death_rate = death_rate.dropna()
death_rate.describe() #fins 2021
```

```python
death_rate.dtypes
```

```python
#afegim la columna death.Rate
merged_oecd = pd.merge(pm25oecd, death_rate, on=['Entity', 'Code', 'Year'], how='left')
merged_oecd = merged_oecd.dropna(subset=['Year', 'Death.Rate'])
merged_oecd['Year'] = merged_oecd['Year'].astype(int)
merged_oecd = merged_oecd.drop_duplicates(subset=['Code', 'Year'])
```

```python
merged_oecd[merged_oecd['Year']==2020]
```

```python
merged_oecd[merged_oecd['Year']==2021]
```

```python
merged_oecd = merged_oecd.fillna('NA') 

merged_oecd[merged_oecd['Code']=='POL']
```

```python
pol = pd.read_csv('points_from_polygons.csv')
pol = pol[['lng', 'lat', 'Code']]

map_oecd = pd.merge(merged_oecd, pol, on=['Code'], how='left')
#map_oecd = map_oecd.drop_duplicates(subset=['Code', 'Year'])
map_oecd[map_oecd['Code']=='POL']
```

```python
excel_df = pd.read_excel('worldbank.PM25.xls', sheet_name='Data', skiprows=3)

pm25wb = excel_df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                        var_name='Year', value_name='PM2.5')

pm25wb = pm25wb[['Country Name', 'Country Code', 'Year', 'PM2.5']]
pm25wb.rename(columns={'Country Name': 'Entity', 'Country Code': 'Code'}, inplace=True)
pm25wb['Year'] = pm25wb['Year'].astype(int)

pm25wb
```
